````python
#Shapiro-Wilk Normality Test for Activity Intensity & Calories Columns
columns = ['very_active_minutes',
           'fairly_active_minutes',
           'lightly_active_minutes',
           'sedentary_minutes',
           'calories']

for col in columns:
    data = df_daily_activity[col].dropna()
    
    stat, p = shapiro(data)
    print(f'Column: {col}')
    print(f'Statistics={stat:.3f}, p={p:.3f}')
    if p > 0.05:
        print('Data is normally distributed\n')
    else:
        print('Data is not normally distributed\n')
````

<style>
  p {
    font-family: Arial, sans-serif;
    font-size: 13px;
  }
</style>

<p>
  Conclusion: All columns tested with the Shapiro-Wilk Normality test have p-values &lt; 0.05, indicating that none of the five columns have normally distributed data. Therefore, Spearman correlation should be used for correlation testing.
</p>

````python
#Spearman Correlation of Activity Intensity vs Calories
columns = ['very_active_minutes',
           'fairly_active_minutes',
           'lightly_active_minutes',
           'sedentary_minutes',
           'calories']

spearman_corr = df_daily_activity[columns].corr(method='spearman')
spearman_corr.round(2)
````

````python
#Spearman Correlation Heatmap of Activity Intensity vs Calories
columns = ['very_active_minutes',
           'fairly_active_minutes',
           'lightly_active_minutes',
           'sedentary_minutes',
           'calories']

spearman_corr = df_daily_activity[columns].corr(method='spearman').round(2)

plt.figure()
sns.heatmap(spearman_corr,
            annot=True,
            cmap="pink",
            vmin=-1,
            vmax=1,
            linewidths=0.5,
            fmt=".2f",
            annot_kws={"fontsize": 10})

plt.title('Spearman Correlation Heatmap', 
          fontsize=14, 
          fontweight='bold')

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
````

````python
#Average Minutes of Activity Intensity by Weekday/Weekend
df_daily_activity['activity_date'] = pd.to_datetime(df_daily_activity['activity_date'])

df_daily_activity['day_type'] = df_daily_activity['activity_date'].dt.dayofweek.apply(
    lambda x: 'Weekend' if x >= 5 else 'Weekday')

avg_activity = df_daily_activity.groupby('day_type')[['very_active_minutes', 
                                                      'fairly_active_minutes', 
                                                      'lightly_active_minutes', 
                                                      'sedentary_minutes']].mean().round(2)
avg_activity.columns = ['avg_very_active_minutes', 'avg_fairly_active_minutes', 'avg_lightly_active_minutes', 'avg_sedentary_minutes']
avg_activity
````

````python
#Distribution of Total Steps Categories
def classify_steps(steps):
    if steps < 5000:
        return 'Sedentary'
    elif steps < 7500:
        return 'Low Active'
    elif steps < 10000:
        return 'Somewhat Active'
    elif steps < 12500:
        return 'Active'
    else:
        return 'Highly Active'
    
df_daily_activity['step_category'] = df_daily_activity['total_steps'].apply(classify_steps)

steps_counts = df_daily_activity['step_category'].value_counts().sort_index()

plt.rcParams['font.family'] = 'Arial'
plt.figure()
colors = ['#ffccd9', '#ff99b3', '#ff6699', '#ff3366', '#cc0044'] 

plt.pie(steps_counts, 
        labels=steps_counts.index, 
        autopct='%1.2f%%', 
        startangle=140, 
        colors=colors)

plt.title('Distribution of Daily Step Categories', 
          fontweight='bold', 
          fontsize=16)
plt.axis('equal')  
plt.show()
````

<style>

    font-family: Arial, sans-serif;

  }
</style>

<p>
  Conclusion: Based on the daily steps classification using Tudor-Locke's (2004) theory, it can be concluded that the majority of users have not yet reached the recommended daily activity targets, with many leading a sedentary lifestyle.
</p>

````python
#Distribution of Sleep Categories
def classify_sleep(minutes):
    if minutes < 360:
        return 'Short (<6h)'
    elif minutes < 420:
        return 'Borderline (6–7h)'
    elif minutes <= 540:
        return 'Recommended (7–9h)'
    else:
        return 'Long (>9h)'

df_sleep_day['sleep_category'] = df_sleep_day['total_minutes_asleep'].apply(classify_sleep)

sleep_counts = df_sleep_day['sleep_category'].value_counts().reindex(['Short (<6h)', 'Borderline (6–7h)', 'Recommended (7–9h)', 'Long (>9h)'])

plt.rcParams['font.family'] = 'Arial'
plt.figure()

bars = plt.bar(sleep_counts.index, 
               sleep_counts.values, 
               color=['#ff3366'])

plt.title('Distribution of Sleep Categories', 
          fontweight='bold', 
          fontsize=16)

plt.ylabel('Number of Records')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval}', ha='center', fontweight='bold')

plt.show()
````

````python
#Check Matching id and date Entries Across Fitbit Datasets
df_daily_activity['activity_date'] = pd.to_datetime(df_daily_activity['activity_date'])
df_sleep_day['sleep_datetime'] = pd.to_datetime(df_sleep_day['sleep_datetime'])
df_weight_log_info['date'] = pd.to_datetime(df_weight_log_info['date'])

df_daily_activity['activity_date'] = df_daily_activity['activity_date'].dt.date
df_sleep_day['sleep_date'] = df_sleep_day['sleep_datetime'].dt.date
df_weight_log_info['weight_date'] = df_weight_log_info['date'].dt.date

#1. Check Overlaps Between All Three Datasets
common_users_all = (set(df_daily_activity['id']) &
                    set(df_sleep_day['id']) &
                    set(df_weight_log_info['id']))

daily_3 = df_daily_activity[df_daily_activity['id'].isin(common_users_all)]
sleep_3 = df_sleep_day[df_sleep_day['id'].isin(common_users_all)]
weight_3 = df_weight_log_info[df_weight_log_info['id'].isin(common_users_all)]

daily_sleep_merged = pd.merge(daily_3, 
                              sleep_3,
                              left_on=['id', 'activity_date'],
                              right_on=['id', 'sleep_date'],
                              how='inner')

daily_sleep_weight_merged = pd.merge(daily_sleep_merged, 
                                     weight_3,
                                     left_on=['id', 'activity_date'],
                                     right_on=['id', 'weight_date'],
                                     how='inner')

print("Overlap Across All Three Datasets")
print(f"Total matching rows: {len(daily_sleep_weight_merged)}")
print(f"Total unique users: {daily_sleep_weight_merged['id'].nunique()}")

#2. Check overlaps between DAILY ACTIVITY and SLEEP DAY only
common_users_2 = (set(df_daily_activity['id']) & set(df_sleep_day['id']))

daily_2 = df_daily_activity[df_daily_activity['id'].isin(common_users_2)]
sleep_2 = df_sleep_day[df_sleep_day['id'].isin(common_users_2)]

activity_sleep_merged = pd.merge(daily_2, 
                                 sleep_2,
                                 left_on=['id', 'activity_date'],
                                 right_on=['id', 'sleep_date'],
                                 how='inner')

print("\nOverlap Between DAILY ACTIVITY & SLEEP DAY")
print(f"Total matching rows: {len(activity_sleep_merged)}")
print(f"Total unique users: {activity_sleep_merged['id'].nunique()}")
````

````python
#The Impact of Steps Categories and Sleep Duration Categories on Calories
def classify_steps(steps):
    if steps < 5000:
        return 'Sedentary'
    elif steps < 7500:
        return 'Low Active'
    elif steps < 10000:
        return 'Somewhat Active'
    elif steps < 12500:
        return 'Active'
    else:
        return 'Highly Active'
    
def classify_sleep(minutes):
    if minutes < 360:
        return 'Short (<6h)'
    elif minutes < 420:
        return 'Borderline (6–7h)'
    elif minutes <= 540:
        return 'Recommended (7–9h)'
    else:
        return 'Long (>9h)'

activity_sleep_merged['step_category'] = activity_sleep_merged['total_steps'].apply(classify_steps)
activity_sleep_merged['sleep_category'] = activity_sleep_merged['total_minutes_asleep'].apply(classify_sleep)

summary = activity_sleep_merged.groupby(['step_category', 'sleep_category'])['calories'].mean().unstack()
summary.round(2)
````

<style>
  p {
    font-family: Arial, sans-serif;
    font-size: 13px;
  }
</style>

<p>
  Conclusion: The highest calories burned were observed in users with Long (&gt;9 hours) and Recommended (7–9 hours) sleep durations, particularly among those with higher step counts. Meanwhile, the Sedentary group burned the fewest calories across all sleep categories. This indicates that the more active a user is, the more calories they tend to burn—especially when sleep duration is sufficient or extended. However, in the Short sleep group (&lt;6 hours), calorie burn was relatively low across all activity levels. This suggests that both activity level and sleep duration are important factors influencing calorie expenditure.
</p>
